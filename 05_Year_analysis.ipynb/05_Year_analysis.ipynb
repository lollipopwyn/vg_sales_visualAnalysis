```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video_game_df = pd.read_csv("../00_data/vgsales.csv", encoding="")
video_game_df
```

```python

#지역별 연도 그룹핑한 평균
jp_game = video_game_df.groupby("Year")['JP_Sales'].mean()
na_game = video_game_df.groupby("Year")['NA_Sales'].mean()
eu_game = video_game_df.groupby("Year")['EU_Sales'].mean()
ot_game = video_game_df.groupby("Year")['Other_Sales'].mean()
gl_game = video_game_df.groupby("Year")['Global_Sales'].mean()
year = video_game_df.groupby("Year")["Year"].mean()
all_region_df = pd.concat([year, jp_game, na_game, eu_game, ot_game, gl_game], axis = 1)

all_region_df

```

```python
plt.figure(figsize=(10,6))
plt.plot(all_region_df["Year"], all_region_df['JP_Sales'], label="JP_Sales", color="blue")
plt.plot(all_region_df["Year"], all_region_df['NA_Sales'], label="NA_Sales", color="red")
plt.plot(all_region_df["Year"], all_region_df['EU_Sales'], label="EU_Sales", color="green")
plt.plot(all_region_df["Year"], all_region_df['Other_Sales'], label="Other_Sales", color="black")
plt.plot(all_region_df["Year"], all_region_df['Global_Sales'], label="Global_Sales", color="brown")

plt.rc("font", family="Malgun Gothic")
plt.title("연도별 지역 평균매출 비교하기", fontsize= 16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Sales", fontsize= 12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 위의 코드를 간소화
region_sales = video_game_df.groupby("Year")[["Year","JP_Sales",'NA_Sales','EU_Sales','Other_Sales','Global_Sales']].mean()
region_sales
```

```python
#위의 코드를 간소화
plt.figure(figsize=(10,6))

regions ={
    'JP_Sales':'blue',
    'NA_Sales':'red',
    'EU_Sales':'green',
    'Other_Sales':'black',
    'Global_Sales':'brown'
}
for region, color in regions.items():
    plt.plot(region_sales["Year"], region_sales[region], label=region, color=color)


plt.rc("font", family="Malgun Gothic")
plt.title("연도별 지역 평균매출 비교하기", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python

# 지역별 연도그룹핑 후 PC와 콘솔 판매 추이 비교

console_game_df = video_game_df.copy()
pc_game_df = video_game_df.copy()


console_game_df.loc[console_game_df['Platform'] == 'PC',['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]=0
console_game_df.loc[console_game_df['Platform'] != 'PC', 'Platform'] = 'Console'

console_region_sales = console_game_df.groupby("Year")[["Year","JP_Sales",'NA_Sales','EU_Sales','Other_Sales','Global_Sales']].mean()


pc_game_df.loc[pc_game_df['Platform'] != 'PC',['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]=0

pc_region_sales = pc_game_df.groupby("Year")[["Year","JP_Sales",'NA_Sales','EU_Sales','Other_Sales','Global_Sales']].mean()




fig, axs = plt.subplots(2,2, figsize=(10,6)) # 1*2 서브플롯을 사용하려고 했으나 2*2서브플롯 배열이상으로 생성해야함

regions_NA_EU={
    'EU_Sales':'green',
    'NA_Sales':'red',
    'Global_Sales':'brown'
}
regions_JP_Other={
    
    'JP_Sales':'blue',
    'Other_Sales':'black',
    'Global_Sales':'brown'
}
for region, color in regions_NA_EU.items():
    axs[0, 0].plot(console_region_sales["Year"], console_region_sales[region], label=region, color=color)

for region, color in regions_NA_EU.items():
    axs[0, 1].plot(pc_region_sales["Year"], pc_region_sales[region], label=region, color=color)


for region, color in regions_JP_Other.items():
    axs[1, 0].plot(console_region_sales["Year"], console_region_sales[region], label=region, color=color)

for region, color in regions_JP_Other.items():
    axs[1, 1].plot(pc_region_sales["Year"], pc_region_sales[region], label=region, color=color)


axs[0,0].set_title("연도별 콘솔 게임 북미,유럽 평균매출", fontsize=16)
axs[0,0].set_xlabel("Date", fontsize=12)
axs[0,0].set_ylabel("Sales", fontsize=12)
axs[0,0].tick_params(axis='x', rotation=45)


axs[0,1].set_title("연도별 PC 게임 북미,유럽 평균매출 ", fontsize=16)
axs[0,1].set_xlabel("Date", fontsize=12)
axs[0,1].set_ylabel("Sales", fontsize=12)
axs[0,1].tick_params(axis='x', rotation=45)

axs[1,0].set_title("연도별 콘솔 게임 일본,그밖에 평균매출 ", fontsize=16)
axs[1,0].set_xlabel("Date", fontsize=12)
axs[1,0].set_ylabel("Sales", fontsize=12)
axs[1,0].tick_params(axis='x', rotation=45)

axs[1,1].set_title("연도별 PC 게임 일본,그밖에 평균매출 ", fontsize=16)
axs[1,1].set_xlabel("Date", fontsize=12)
axs[1,1].set_ylabel("Sales", fontsize=12)
axs[1,1].tick_params(axis='x', rotation=45)


#서브플롯 간의 간격조정
plt.subplots_adjust(hspace=0.6, wspace=0.4)
```
